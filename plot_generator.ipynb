```python
import pandas as pd
import matplotlib.pyplot as plt
import os
from scipy.signal import savgol_filter

# read in the datafiles
def main():
    # I'm going to use polynomial 3, window size 13
    directory = os.fsencode("SG_folder_tests/results_SG_size_271")
    for file in os.listdir(directory):
        filename = os.fsdecode(file)
        if filename.endswith(".txt"):
            file_name_split = filename.split("_")
            plot_title = "Dataset:" + file_name_split[0]
            #+ " Window Size: "+ file_name_split[1]
            df = pd.read_csv("SG_folder_tests/results_SG_size_271"+ "/" + filename, sep="\t")
            #filtered_discrim = savgol_filter(df["Discriminant_Value"], 13, 3)
            plt.plot(df["Start_Pos"], df["Discriminant_Value"])
            plt.ylim(-0.4,0.4)
            plt.title(plot_title)
            #plot_path = "SG_filter_poly3_width13_plot_images_6_30_20/" + file_name_split[0]+"_"+file_name_split[1]+".png"
            #plt.savefig(plot_path, dpi = 300)
            plt.show()
```

```python
main()
```
